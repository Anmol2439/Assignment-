```python
#5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate a synthetic dataset with correlated features
np.random.seed(42)  # For reproducibility
n_samples = 100

# Create correlated features
x = np.random.randn(n_samples)
y = 2 * x + np.random.randn(n_samples) * 0.1  # y is correlated with x
z = 0.5 * x + 0.5 * y + np.random.randn(n_samples) * 0.1  # z is correlated with both x and y

# Combine into a DataFrame
data = pd.DataFrame({'Feature X': x, 'Feature Y': y, 'Feature Z': z})

# Calculate the correlation matrix
correlation_matrix = data.corr()

# Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', square=True, cbar_kws={"shrink": .8})

# Add title
plt.title('Correlation Matrix Heatmap')

# Display the plot
plt.show()
```

```python
#3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and days.
#np.random.seed (20)
#data = {
#'Month': np.random.choice (['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
#'Day': np.random.choice(range (1, 31), 100),
#'Sales': np.random.randint(1000, 5000, 100)
#}
#df = pd.DataFrame (data)
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(20)

# Generate the data
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}

# Create a DataFrame
df = pd.DataFrame(data)

# Create a pivot table for the heatmap
pivot_table = df.pivot_table(values='Sales', index='Day', columns='Month', aggfunc='mean')

# Create a heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(pivot_table, annot=True, fmt=".1f", cmap='YlGnBu', cbar_kws={'label': 'Average Sales'})

# Adding labels and title
plt.xlabel('Month')
plt.ylabel('Day of the Month')
plt.title('Heatmap of Sales Variation Across Different Months and Days')

# Show the plot
plt.show()
```

```python
#4. Using the given x and y data, generate a 3D surface plot to visualize the function z = sin(x+y
#x = np.linspace (-5, 5, 100)
#y = np.linspace (-5, 5, 100)
#x, y = np.meshgrid(x, y)
#z = np.sin(np.sqrt(x**2 + y**2))
#data = {
#'X': x.flatten(),
#'Y': y.flatten(),
#'Z': z.flatten()
#}
#df = pd.DataFrame (data)
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate the data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Create a DataFrame
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Create a 3D surface plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plotting the surface
ax.plot_surface(df['X'].values.reshape(100, 100), df['Y'].values.reshape(100, 100),
                 df['Z'].values.reshape(100, 100), cmap='viridis')

# Adding labels and title
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis (sin(sqrt(x + y)))')
ax.set_title('3D Surface Plot of z = sin(sqrt(x + y))')

# Show the plot
plt.show()
```

```python
#1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.
# Import necessary libraries
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Set output to notebook (can also use output_file('filename.html') for standalone HTML)
output_notebook()

# Generate data for the sine wave
x = np.linspace(0, 10, 100)  # x values from 0 to 10
y = np.sin(x)                 # y values as the sine of x

# Create a new plot
p = figure(title="Sine Wave", x_axis_label='X', y_axis_label='sin(X)', width=800, height=400)

# Add a line renderer for the sine wave
p.line(x, y, legend_label="sin(x)", line_width=2, line_color="blue")

# Show the plot
show(p)
```

```python
#2. Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.
# Import necessary libraries
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Set output to notebook (can also use output_file('filename.html') for standalone HTML)
output_notebook()

# Generate random data
np.random.seed(42)  # For reproducibility
n = 100  # Number of points
x = np.random.rand(n) * 100  # Random x values
y = np.random.rand(n) * 100  # Random y values
sizes = np.random.randint(10, 100, n)  # Random sizes for markers
colors = np.random.choice(['red', 'green', 'blue', 'orange', 'purple'], n)  # Random colors

# Create a DataFrame to organize the data
data = {
    'x': x,
    'y': y,
    'sizes': sizes,
    'colors': colors
}
df = pd.DataFrame(data)

# Create a scatter plot
p = figure(title="Scatter Plot with Varying Sizes and Colors", x_axis_label='X', y_axis_label='Y', width=800, height=400)

# Add scatter renderer
p.scatter('x', 'y', size='sizes', color='colors', legend_field='colors', source=df, fill_alpha=0.6)

# Show the plot
show(p)
```

```python
#3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
#fruits = ['Apples', 'Oranges', 'Bananas', 'Pears'] counts = [20, 25, 30, 35]
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import pandas as pd

# Enable output in a Jupyter notebook (comment this line if not in Jupyter)
output_notebook()

# Define the dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a DataFrame
df = pd.DataFrame({'Fruits': fruits, 'Counts': counts})

# Create a Bokeh figure
p = figure(x_range=df['Fruits'], height=350, title="Fruit Counts",
           toolbar_location=None, tools="")

# Create the bar chart
p.vbar(x='Fruits', top='Counts', width=0.9, source=df, color='blue')

# Formatting the plot
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.yaxis.axis_label = "Counts"
p.xaxis.axis_label = "Fruits"
p.xaxis.major_label_orientation = "vertical"

# Show the plot
show(p)

```

```python
#4 Create a Bokeh histogram to visualize the distribution of the given data.
#data_hist = np.random.randn(1000) hist, edges = np.histogram(data_hist, bins=30)
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Enable output in a Jupyter notebook (comment this line if not in Jupyter)
output_notebook()

# Generate random data
data_hist = np.random.randn(1000)

# Compute histogram
hist, edges = np.histogram(data_hist, bins=30)

# Create a Bokeh figure
p = figure(title="Histogram of Random Data",
           x_axis_label='Value',
           y_axis_label='Frequency',
           tools="")

# Create the histogram bars
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color='blue', line_color='black')

# Show the plot
show(p)
```
